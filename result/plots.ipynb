```python
import matplotlib.pyplot as plt
import numpy as np
import csv

plt.style.use('seaborn-darkgrid')
```

```python
with open("walker_run_finetune/train.csv") as f:
    reader = csv.reader(f)
    next(reader)
    train = list(reader)
```

```python
episodes = [float(row[1]) for row in train]
rewards = np.array([float(row[3]) for row in train])

plt.plot(episodes, rewards, linewidth=1, color='red')

plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')
plt.xlim(-10, 310)
plt.ylim(-30, 750)

plt.xlabel("Episodes")
plt.ylabel("Average Reward")
plt.title("Walker Run Finetune")
plt.savefig("walker_run_finetune.png")
plt.show()
```
